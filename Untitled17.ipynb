```python
import pandas as pd

# Đọc dữ liệu từ tệp lazada.books.csv
lazada_data = pd.read_csv('/content/lazada.books.csv')

# Loại bỏ giá trị "Off" và thêm dấu "-" đằng trước trong cột "Giảm giá" của lazada
lazada_data['Giảm giá'] = lazada_data['Giảm giá'].str.replace(' Off', '').str.replace(',', '').str.replace('₫', '').str.replace('%', '')
lazada_data['Giảm giá'] = '-' + lazada_data['Giảm giá']

# Thêm ký tự '%' đằng sau cột "Giảm giá" của lazada
lazada_data['Giảm giá'] = lazada_data['Giảm giá'] + '%'

# Chuyển đổi cột "Giảm giá" sang kiểu số float
lazada_data['Giảm giá'] = lazada_data['Giảm giá'].str.replace(r'[^0-9.-]', '', regex=True).astype(float)

# Xuất ra một tệp CSV mới cho Lazada
lazada_data.to_csv('/content/lazada_modified.csv', index=False, encoding='utf-8-sig')

# Hiển thị dữ liệu đã lọc của lazada
print(lazada_data['Giảm giá'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Tạo một DataFrame từ dữ liệu CSV
data = pd.read_csv('tiki.books.csv')

# Lấy cột "Phần trăm giảm giá"
percent_discount = data['Giảm giá'].str.replace('%', '').astype(float)

# Tạo một biểu đồ đường
plt.figure(figsize=(12, 6))  # Kích thước biểu đồ
plt.plot(percent_discount, marker='o', linestyle='-')

# Đặt tiêu đề và nhãn cho biểu đồ
plt.title('Biểu đồ phân tích Phần trăm giảm sàn Tiki')
plt.xlabel('Sách')
plt.ylabel('Phần trăm giảm giá')

# Hiển thị giá trị trục y dưới dạng phần trăm
plt.gca().set_yticklabels(['{:.0f}%'.format(x) for x in plt.gca().get_yticks()])

# Hiển thị biểu đồ
plt.grid(True)  # Hiển thị lưới
plt.xticks(rotation=90)  # Xoay nhãn trục x để dễ đọc
plt.tight_layout()  # Đảm bảo rằng nhãn không bị trùng lên nhau

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('/content/lazada.books.csv')

# Xử lý dữ liệu trong cột Giá tiền
data['Giá tiền'] = data['Giá tiền'].str.replace('₫', '').str.replace(',', '').astype(float)

# Loại bỏ ký tự "%" và "Off" khỏi cột Giảm giá và chuyển đổi thành kiểu số
data['Giảm giá'] = data['Giảm giá'].str.replace('%', '').str.replace('Off', '').astype(float)

# Nhóm dữ liệu theo Giảm giá và tính giá trung bình cho mỗi nhóm
mean_discount = data.groupby('Giảm giá')['Giá tiền'].mean().reset_index()

# Tạo biểu đồ đường
plt.figure(figsize=(12, 6))
plt.plot(mean_discount['Giảm giá'], mean_discount['Giá tiền'], marker='o', linestyle='-')
plt.xlabel('Phần trăm giảm giá')
plt.ylabel('Giá trung bình')
plt.title('Biểu đồ đường phân tích giảm giá sàn Lazada')
plt.grid(True)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ hai tệp CSV
lazada_data = pd.read_csv('/content/lazada.books.csv')
tiki_data = pd.read_csv('/content/tiki.books.csv')

# Loại bỏ các giá trị không hợp lệ trong cột "Giảm giá" của cả hai tệp
lazada_data = lazada_data[~lazada_data['Giảm giá'].str.contains('Off')]
tiki_data = tiki_data[~tiki_data['Giảm giá'].str.contains('Off')]

# Chuyển đổi cột "Giảm giá" thành kiểu float
lazada_data['Giảm giá'] = lazada_data['Giảm giá'].str.replace('%', '').str.replace(',', '').astype(float)
tiki_data['Giảm giá'] = tiki_data['Giảm giá'].str.replace('%', '').str.replace(',', '').astype(float)

# Tạo biểu đồ dạng đường cho lazada_data
plt.figure(figsize=(12, 6))
plt.plot(lazada_data['Giảm giá'], label='Lazada', color='skyblue', marker='o')
plt.plot(tiki_data['Giảm giá'], label='Tiki', color='lightgreen', marker='o')

plt.xlabel('Số sách')
plt.ylabel('Phần trăm giảm giá')
plt.title('Biểu đồ đường phân tích giảm giá theo số sách')

plt.legend()
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('/content/books.csv')

# Trích xuất dữ liệu số từ cột "Giảm giá" và chuyển đổi nó thành kiểu float
data['Giảm giá'] = data['Giảm giá'].str.replace('%', '').str.replace(',', '').astype(float)

# Thống kê mô tả cơ bản
discount_stats = data['Giảm giá'].describe()

# Vẽ biểu đồ phân phối
plt.figure(figsize=(8, 6))
plt.hist(data['Giảm giá'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Phần trăm giảm giá')
plt.ylabel('Số lượng sách')
plt.title('Phân phối phần trăm giảm giá')
plt.grid(True)

# Hiển thị thống kê mô tả
print("Thống kê mô tả về phần trăm giảm giá:")
print(discount_stats)

# Hiển thị biểu đồ
plt.show()
```
